```python
# 케라스 개발 과정:
# 데이터 생성 > 모델 구축 > 모델 컴파일 > 모델 학습 > 모델 평가 및 예측 > 모델 저장

# 데이터 생성
# 전체 데이터로부터 train data / validation data / test data 등을 생성함
# train data : 학습에 사용되는 데이터, 가중치와 바이어스를 최적화 하기 위해 사용
# validation data : 1 epoch 마다 과적합(overfitting)을 확인하기 위해 사용
# test data : 학습 후에 정확도를 평가하거나 임의의 입력에 대한 결과를 예측하기 위해 사용

# 개발 과정(모델 구축)
# model = Sequential() (입력층 > 은닉층 > 출력층)
# model.add(Flatten(input_shape=(1,)))
#  Faltten : 입력으로 들어오는 다차원 데이터를 1차원으로 정렬하기 위해 사용되는 레이어이며, 입력데이터(차원)의 수를 input_shape과 같이 기술
# model.add(Dense(2, activation = 'sigmoid'))
# model.add(Dense(1, activation = 'sigmoid'))
#  Dense : 각 층의 입력과 출력 사이에 있는 모든 노드가 서로 연결되어 있는 완전 연결층(FC)을 나타내며,첫번째 인자 2,1은 출력 노드 수를 나타냄
#    노드의 활성화 함수는 activation = '...' 형태로 나타내며, 대표적인 활성화 함수는 linear, sigmoid, softmax, relu, tanh 등 데이터에 따라 다름
# model.add(Dense(2, activation = 'sigmoid', input_shape=(1,)))
#    첫 번째 계층에 바로 Dense 계층을 사용하는 경우라면 위와 같이 한 줄로 나타낼 수 있음 > 일반적으로 이렇게 Flatten을 사용하지 않고 한 줄의 코드로 나타냄

# 모델 컴파일
# model.compile() 후에는 model.summay()를 통해서 구축 모델을 확인함
# 예1) model.compile(optimizer = SGD(learning_rate = 0.1), loss = 'mse', matrics = ['accuracy'])
#    최적화 알고리즘 SGD, 학습율 0.1, 오차함수 mse, 매트릭은 accuracy(loss는 기본)
# 예2) model.compile(optimizer = Adam(learning_rate = 1e-4), loss = 'categorical_crossentropy')
#    최적화 알고리즘 Adam, 학습율 1e-4, 오차함수 categorical crossentropy
# 손실함수의 종류는 mse(평균제곱오차), binary_crossentropy(이진분류), categorical_crossentropy(다중클래스분류), sparse_categorical_corssentropy
# 매트릭은 기본적으로 loss만 측정함. 즉 metrics = ['loss'] 임

# 모델 학습
# 손실함수값이 최소가 될때까지 각 층의 가중치와 바이어스를 업데이트
# 예) model.fit(x_train, t_train, epochs = 10, batch_size = 100, verbose = 0, validation_split = 0.2)
#    입력데이터, 정답데이터, epochs(전체데이터를 n번 반복), 배치사이즈(생략가능), validation_split: 학습데이터의 20%를 검증데이터로 활용
#    verbose = 학습 중 손실값, 정확도, 진행상태 등의 출력 형태를 설정함(0, 1, 2)
#    검증데이터가 있다면 validation_data, 없다면 validation_split을 사용함

# 모델 평가
# 예1) model.evaluate(x_data, t_test, epochs = 10, batch_size = 100)
# 예2) model.predict(x_input_data, batch_size = 100)

# 모델 저장
# model.save("") : 저장
# model = tensorflow.keras.models.load_model("") : 불러오기
# 학습된 신경망 구조를 저장해 놓으면 다양한 테스트 데이터에 대해 재학습 없이 지속적으로 사용할 수 있음
```

```python
# 예제
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD

import numpy as np

x_data = np.array([1,2,3,4,5,6])
t_data = np.array([3,4,5,6,7,8])

model = Sequential()
model.add(Dense(1,input_shape=(1,), activation = 'linear'))
model.compile(optimizer = SGD(learning_rate = 1e-2), loss = 'mse')
model.summary()
```

```python
hist = model.fit(x_data, t_data, epochs = 1000)
```

```python
result = model.predict(np.array([-3.1, 3.0, 3.5, 15.0, 20.1]))
print(result)
```
